```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import holoviews as hv
hv.extension('bokeh')
```

```python
train_df = pd.read_csv(filepath_or_buffer='../data/train.csv')
test_df = pd.read_csv(filepath_or_buffer='../data/test.csv')
```

### Exploring training dataset

```python
fig,ax = plt.subplots(ncols=3, figsize=(9,3))
sns.histplot(data=train_df.query('phase=="Gesture"').sequence_id.value_counts(), bins=50, ax=ax[0]);
sns.histplot(train_df.query('behavior=="Hand at target location"').sequence_id.value_counts(), bins=50, ax=ax[1]);
sns.histplot(train_df.query('behavior=="Moves hand to target location" or behavior=="Relaxes and moves hand to target location"').sequence_id.value_counts(), bins=50, ax=ax[2]);
ax[0].set_title('Gesture Phase')
ax[1].set_title('Pause Phase')
ax[2].set_title('Transition Phase')
ax[0].set_xlabel('number of frames')
ax[1].set_xlabel('number of frames')
ax[2].set_xlabel('number of frames')
fig.tight_layout()
```
